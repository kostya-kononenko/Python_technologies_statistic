```python
import pandas as pd
from pprint import pprint
from collections import Counter
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('parse/vacancies.csv')
df.head()
```

```python
df1 = df['technologies'].tolist()
df2 = str(df1).replace(" ", "").replace("[", "").replace("]", "").replace("'", "").replace('"', "").replace("/", ",").replace("FastApi", "FastAPI").replace("python", "Python").replace("RESTAPI", "DjangoRestFramework").replace("Python3", "Python")
df3 = df2.split(",")
df4 = Counter(df3)
pprint(df4)
```

```python
df5 = pd.DataFrame({'name':df4.keys(), 'value':df4.values()})
df6 = pd.DataFrame(df5)
```

```python
df6.sort_values(by='value', ascending=False).iloc[0:30]
```

```python
df7 = df6.drop([16, 47, 28, 97, 46])
```

```python
df8 = df7.sort_values(by='value', ascending=False)
df9 = df8.iloc[0:30]
```

```python
plt.figure()
ax = df9.plot(kind='bar',
    x="name", 
             y="value",
             title="The most popular technologies in the Python",
             xlabel = "Technologies",
             ylabel = "Count",
             figsize=(20,10),
             fontsize='large',
             color = "green",
             )
ax.bar_label(ax.containers[0], label_type='edge')
```
